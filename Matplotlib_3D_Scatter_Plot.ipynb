## *3D Scatter Plot* dari Array

```python
# Menyiapkan Data X, Y, dan Z
X = np.array([5,  3, 5,  2, 7,  1, 9,  6,  10,  2, 2,  4, 6,  8, 3, 11])
Y = np.array([15, 3, 7, 12, 5,  3, 20, 11, 18,  1, 5,  9, 7, 20, 3, 10])
Z = np.array([ 2, 3, 9,  5, 9, 11,  2,  5,  2, 15, 8, 11, 4, 16, 8, 11])

# Membuat 3D Scatter Plot
fig = plt.figure(figsize = (8,6))                      # Ukuran Gambar (dalam inci)
ax = fig.add_subplot(projection = '3d')

ax.scatter3D(X, Y, Z)

ax.set_title("3D Scatter Plot dari Array")             # Judul Gambar

ax.set_xlabel("X")                                     # Keterangan untuk Sumbu X
ax.set_ylabel("Y")                                     # Keterangan untuk Sumbu Y
ax.set_zlabel("Z")                                     # Keterangan untuk Sumbu Z

ax.invert_xaxis()                                      # Membalikkan Sumbu X

ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu X Menjadi Integer
ax.yaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu Y Menjadi Integer
ax.zaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu Z Menjadi Integer

plt.show()
```

### Mengatur Orientasi/Sudut Pandang *3D Scatter Plot*

```python
# Menyiapkan Data X, Y, dan Z
X = np.array([5,  3, 5,  2, 7,  1, 9,  6,  10,  2, 2,  4, 6,  8, 3, 11])
Y = np.array([15, 3, 7, 12, 5,  3, 20, 11, 18,  1, 5,  9, 7, 20, 3, 10])
Z = np.array([ 2, 3, 9,  5, 9, 11,  2,  5,  2, 15, 8, 11, 4, 16, 8, 11])

# Membuat 3D Scatter Plot
fig = plt.figure(figsize = (8,6))                      # Ukuran Gambar (dalam inci)
ax = fig.add_subplot(projection = '3d')

ax.scatter3D(X, Y, Z)

ax.view_init(elev = 10, azim = 10, roll = 0)           # Mengatur Orientasi/Sudut Pandang

ax.set_title("3D Scatter Plot dari Array")             # Judul Gambar

ax.set_xlabel("X")                                     # Keterangan untuk Sumbu X
ax.set_ylabel("Y")                                     # Keterangan untuk Sumbu Y
ax.set_zlabel("Z")                                     # Keterangan untuk Sumbu Z

ax.invert_xaxis()                                      # Membalikkan Sumbu X
ax.invert_yaxis()                                      # Membalikkan Sumbu y

ax.xaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu X Menjadi Integer
ax.yaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu Y Menjadi Integer
ax.zaxis.set_major_locator(MaxNLocator(integer=True))  # Ticks Sumbu Z Menjadi Integer

plt.show()
```
